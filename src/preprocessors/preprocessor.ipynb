```python
import os
import pandas as pd

dataset_dir = "/Users/2210a00076/Coding/cl_bert_kt/datasets/real_dataset"

question_meta_path = os.path.join(dataset_dir, "question_meta.csv")
question_meta_df = pd.read_csv(
    question_meta_path,
    encoding="utf-8",
    sep='|'
)
print("question_meta_df.head(): ", question_meta_df.head())
```

```python
# clean the html tags of question_meta_df['question_body']
import re
def clean_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)

question_meta_df['question_body'] = question_meta_df['question_body'].apply(clean_html_tags)

question_meta_df['question_body']
```

```python
# replace "&nbsp;"" to " " in question_meta_df['question_body']
question_meta_df['question_body'] = question_meta_df['question_body'].str.replace("&nbsp;", " ")

question_meta_df['question_body']
```

```python
pre_question_meta_df = question_meta_df[['question_id', 'question_body']]

pre_question_meta_df
```

# Question diff

```python
question_df = pd.read_csv(
    "/Users/2210a00076/Coding/cl_bert_kt/datasets/real_dataset/questions.csv",
    encoding="utf-8",
    sep='|'
)
question_df.head()
```

```python
# for diff
diff_df = pd.read_csv(
    "/Users/2210a00076/Coding/cl_bert_kt/datasets/real_dataset/questions_acc.csv",
    encoding="utf-8",
    sep='|'
)
diff_df.head()
```

```python
diff_df['diff'] = diff_df['n_correct'] / diff_df['n_total']

diff_df.head()
```

```python
diff_df['question_id'] = diff_df['question_id'].replace("Q", "", regex=True)

# if question_id is "PAD", then erase row
diff_df = diff_df[diff_df['question_id'] != "PAD"]

# diff_df['question_id] to int
diff_df['question_id'] = diff_df['question_id'].astype(int)

diff_df.head()
```

```python
# loop for question_id and question_body

question_ids = []
difficulties = []
question_texts = []


for row in question_df.iterrows():
    try:
        question_id = row[1]['question_id']

        print("question_id: ", question_id)

        difficulty = diff_df[
            diff_df['question_id'] == question_id
            ]['diff'].values[0]
        
        print("difficuty", difficulty)

        question_text = pre_question_meta_df[
            pre_question_meta_df['question_id'] == question_id
            ]['question_body'].values[0]

        question_ids.append(question_id)
        difficulties.append(difficulty)
        question_texts.append(question_text)
    except:
        print("error: ", row[1]['question_id'])
        continue


# make question_df
question_df = pd.DataFrame({
    'question_id': question_ids,
    'difficulty': difficulties,
    'question_text': question_texts
})

question_df.head()
```

```python
question_df.to_csv("question_df.csv", sep="|", encoding="utf-8", index=False)
```
